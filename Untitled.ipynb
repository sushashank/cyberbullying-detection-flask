```python
import pandas as pd

# Load your dataset
data = pd.read_csv('data/cyberbullying_tweets(ML).csv')
print(data.head())
print(f"Dataset shape: {data.shape}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

# Ensure the data has the expected structure
print(data.columns)

# Preprocess and split the data
X = data['tweet_text'].astype(str)  # Replace 'tweet' with your actual column name
y = data['cyberbullying_type']  # Replace 'label' with your actual label column name

# Vectorize the text data
vectorizer = TfidfVectorizer(stop_words='english')
X_vectorized = vectorizer.fit_transform(X)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X_vectorized, y, test_size=0.2, random_state=42)

# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Save the model and vectorizer
import pickle
with open('models/vectorizer.pkl', 'wb') as vec_file:
    pickle.dump(vectorizer, vec_file)

with open('models/best_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

print("✅ Model and vectorizer saved successfully!")
```

```python
# Predict using test data
sample_text = ["fuck", "you are amazing", "idiot", "hate you", "lovely day"]
sample_vector = vectorizer.transform(sample_text)
sample_predictions = model.predict(sample_vector)

# Display predictions
for text, prediction in zip(sample_text, sample_predictions):
    result = "Cyberbullying Detected!" if prediction == 1 else "No Cyberbullying Detected!"
    print(f"Text: '{text}' => {result}")
```
